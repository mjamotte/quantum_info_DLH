## From the Fourier transform to number factorization

In the 1990s, Shor developed several algorithms for solving similar problems sharing a common point: revealing hidden regularities. The key discovery that enabled the development of these algorithms was the **Quantum Fourier Transform (QFT)**.

Several problems then benefited from this finding, in the fields of arithmetic and group theory

- Finding the "order of a number modulo N"
- Computing discrete logarithms
- Hidden subgroup search
- **Number factorization**

It is the last application that will interest us here. It is also the most well-known, due to its major implications for cryptography. Indeed, the most widely used encryption protocol (RSA) relies on the fact that it is **impossible for classical computers to factor large numbers in a reasonable time**.

## Factorization

Current encryption protocols rely on the difficulty of factoring large numbers $N=pq$, constructed from the product of two large prime numbers $p$ and $q$. Messages are encrypted with the number $N$, which is public information, and decrypted with the numbers $p$ and $q$, which are kept secret.

Deciphering a message therefore consists of finding the numbers $p$ and $q$ that factorize $N$.

### Side note on "modulos"

In mathematics, the property "p divides N" can be written in several different ways. One of them is the following

\begin{equation*}
    N = 0 \mod p
\end{equation*}

It uses the "modulo" (mod) notation, which conveys the idea that the remainder of dividing $N$ by $p$ is $0$. Other examples,

\begin{align*}
    0 &= 0 \mod 4 \\
    2 &= 2 \mod 4 \\
    4 &= 0 \mod 4 \\
    10 &= 2 \mod 4 \\
\end{align*}

-------------------------------------------------------------------------

The trick behind Shor's algorithm is to find a large number $X$ that is itself divisible by $N$, and that can itself be (easily) factored into a product $X = x_1 x_2$.

We also seek to find such a number $X$ such that the factors $x_1$ and $x_2$ are however *not* divisible by $N$.

If $X$ is divisible by $N$, and $N$ is itself divisible by $p$ and $q$, it follows that $X$ is also divisible by $p$ and $q$. Let us focus on the factor $p$. If $p$ divides $X$, then either $x_1$ or $x_2$ must be divisible by $p$ in turn.

\begin{equation*}
    x_1 x_2 = 0 \mod p \iff (x_1 = 0 \mod p) \quad\text{OR}\quad (x_2 = 0\mod p)
\end{equation*}

In fact, it can be shown that if $x_1$ is the number that $p$ divides, then the number $x_2$ is the one divisible by $q$.

This is very good news: it means that the numbers $x_1$ and $x_2$ each share a common factor with $N$ (it is either $p$ or $q$). And there exists a very fast (classical) method to find a common factor between two numbers: **Euclid's algorithm**.

At this stage, the remaining question is: *How do we find this mysterious number $X$?*

This is where an algorithm developed by Shor becomes useful: Starting from any number $1 < a < N$, Shor is able to find a number $r$ that satisfies:

\begin{equation*}
    a^r = 1 \mod N
\end{equation*}

In mathematics, $r$ is called *the order of $a$ modulo $N$*. In particular, if $r$ is even, it allows constructing the number $X$ with the properties we are looking for:

\begin{equation*}
    \underbrace{(a^r-1)}_{X} = 0 \mod N
    \qquad\text{and}\qquad
    \underbrace{(a^r-1)}_{X} = \underbrace{(a^{r/2}-1)}_{x_1}\underbrace{(a^{r/2}+1)}_{x_2}
\end{equation*}

If $r$ turns out to be odd, one simply needs to try again with another value of $a$.

## Phase estimation

The phase estimation algorithm allows estimating the phase $\theta\in[0,1[$ associated with a given eigenstate of an operator $U$:

\begin{equation*}
    U \ket{k} = e^{2\pi i\theta} \ket{k}
\end{equation*}

The quantum circuit transforms an initial state $\ket{0}^{\otimes 2n}\ket\psi$ into a superposition $\sum_k c_k \ket{k}\ket\psi$ peaked around integers $k$ close to the value $2^{2n}\theta$, encoded on $2n$ bits.

## Finding the order of a number modulo N

Application of the phase estimation algorithm for an operator $U$ implementing multiplication by $a$ (modulo $N$):

\begin{equation*}
    U \ket{k} =
    \begin{cases}
        \ket{ak\mod N} &\text{ if } 0 \leq k < N  \\
        \ket{k}        &\text{ if } N \leq k < 2^n
    \end{cases}
\end{equation*}

Since $U^r=I$, the eigenvalues of $U$ are the roots of unity, $\omega_j = e^{2\pi i j/r}$. The associated eigenstates are:

\begin{equation*}
    \ket{\psi_0} = \frac{\ket{1} + \ket{a} + \ket{a^2} + \cdots + \ket{a^{r-1}}}{\sqrt{r}}
\end{equation*}

\begin{equation*}
    \ket{\psi_j} = \frac{\ket{1} + \omega_j^{-1}\ket{a} + \omega_j^{-2}\ket{a^2} + \cdots + \omega_j^{-r+1}\ket{a^{r-1}}}{\sqrt{r}}
\end{equation*}

Indeed, $U\ket{a} = \ket{a^2}$, etc., which makes these states eigenvectors of $U$:

\begin{equation*}
    U\ket{\psi_0} = \frac{\ket{a} + \ket{a^2} + \ket{a^3} + \cdots + \ket{a^{r}}}{\sqrt{r}} = \ket{\psi_0} 
\end{equation*}

\begin{equation*}
    U\ket{\psi_j} = \frac{\ket{a} + \omega_j^{-1}\ket{a^2} + \omega_j^{-2}\ket{a^3} + \cdots + \omega_j^{-r+1}\ket{a^{r}}}{\sqrt{r}} = \omega_j\ket{\psi_j} 
\end{equation*}

The phase estimation algorithm on a state $\ket{\psi_j}$ allows in principle extracting the associated phase $\theta = j/r$. However, it is impossible for us to prepare the states $\ket{\psi_j}$, because the number $r$ is unknown to us. The trick is to simply run the phase estimation algorithm on the initial state $\ket{1}$, which can be written as the following combination of eigenvectors of $U$:

\begin{equation*}
    \ket{1} = \frac{\ket{\psi_0} + \ket{\psi_1} + \cdots + \ket{\psi_{r-1}}}{\sqrt{r}}
\end{equation*}

The phase estimation algorithm will then return an estimate of the phase $\theta = j/r$ for some random integer $j$.

The procedure is repeated a certain number of times to collect several phase approximations $\theta = j/r$, and a fast classical algorithm is then used to extract the value $r$.

The most expensive part is the implementation of the operations $U^{2^j}$, achievable in $O(n^3)$ operations.
