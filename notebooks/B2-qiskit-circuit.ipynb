## Qu'est-ce qu'un circuit?

Un calcul à exécuter sur processeur quantique se représente typiquement sous la forme d'un circuit. Le circuit est une représentation schématique où les qubits sont représentés par des lignes horizontales interrompues par des *portes*, représentant les opérations effectuées sur ces qubits. 

Le circuit décrit une opération générale sur l'ensemble des qubits, que l'on suppose initialisés dans l'état $\ket 0$, sauf indication contraire. Sa lecture se fait de gauche à droite, les portes modifiant petit à petit l'état quantique du système. L'état final se lit donc à la sortie du circuit (droite). Physiquement, les portes représentent des opérations qui seront réellement effectuées sur l'état quantique du processeur physique. La lecture de gauche à droite est donc interprétable comme une évolution temporelle du système.

Pour extraire une information d'un état quantique, il est nécessaire d'effectuer des *mesures* sur les qubits. Les mesures sont toujours supposées être effectuées sur tous les qubits de l'état final, et donc placées à droite au bout des lignes horizontales. Les mesures permettent de convertir l'information sous une forme interprétable par un humain ou un ordinateur classique, matérialisée par la séquence binaire des résultats de mesure.

## Construction de circuits simples

Création d'un état superposé sur un seul qubit:

```python
from qiskit.circuit import QuantumCircuit
from qiskit.quantum_info import Statevector, Operator 

# 1: Créer un circuit à 1 qubit:
circuit = QuantumCircuit(1)

# 2: Ajouter une porte de Hadamard sur le qubit 0:
circuit.h(0)

# 3: Visualisation du circuit
display(circuit.draw('mpl'))

# 4: Visualisation de la matrice correspondante:
circuit_matrix = Operator.from_circuit(circuit)
display(circuit_matrix.draw('latex'))

# 5: Visualisation de l'état final:
circuit_final_state = Statevector.from_circuit(circuit)
display(circuit_final_state.draw('latex'))
```

Paire de Bell:

```python
from qiskit.circuit import QuantumCircuit
from qiskit.quantum_info import Statevector, Operator 

# 1: Créer un circuit à 2 qubits:
circuit = QuantumCircuit(2)

# 2: Ajouter la porte de Hadamard sur le qubit 0 et un opération CNOT sur 1 controlée par 0:
circuit.h(0)
circuit.cx(0,1)

# 3: Visualisation du circuit
display(circuit.draw('mpl'))

# 4: Visualisation de la matrice correspondante:
circuit_matrix = Operator.from_circuit(circuit)
display(circuit_matrix.draw('latex'))

# 5: Visualisation de l'état final:
circuit_final_state = Statevector.from_circuit(circuit)
display(circuit_final_state.draw('latex'))
```

```python
display(circuit.draw('text'))
```
