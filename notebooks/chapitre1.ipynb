## 1.1: Pourquoi recourir aux ordinateurs et simulateurs quantiques?

Afin de manipuler des données avec un ordinateurs, il faut pouvoir les inscrire dans un registre. Pour un système de particules, il faut donc encoder les informations de chacune d'elles.

Agir sur $N$ qubits se fait avec un nombre d'opérations qui grandit comme $N$ (dans le meilleur cas), tandis qu'un ordinateur classique requiert d'agir sur $2^N$ bits.
