# Bell inequalities

## EPR paradox

<img src="../figures/epr.png" alt="chsh" width="400"/>

Entangled states,

\begin{equation*}
    \ket{\psi_{AB}} = \frac{\ket{00} + \ket{11}}{\sqrt{2}} \sim \frac{\ket{++} + \ket{--}}{\sqrt{2}}
\end{equation*}

lead to surprising correlations, which suggest an instantaneous action at a distance.

If Bob alone performs measurements, he always has a $50\%$ chance of obtaining one of the two possible states ($\ket 0$ or $\ket 1$ in the $Z$ basis, and $\ket +$ or $\ket -$ in the $X$ basis). The measured expectation values are therefore zero:

\begin{equation*}
    \langle Z_B \rangle = \left( \frac{\bra{00} + \bra{11}}{\sqrt{2}} \right) \left( \frac{\ket{00} + (-1) \ket{11}}{\sqrt{2}} \right) = 0
\end{equation*}

\begin{equation*}
    \langle X_B \rangle = \left( \frac{\bra{00} + \bra{11}}{\sqrt{2}} \right) \left( \frac{\ket{01} + \ket{10}}{\sqrt{2}} \right) = 0
\end{equation*}

However, if Alice performs measurements in the $Z$ basis, Bob will observe (once Alice has shared her results with him) that his results are totally correlated with Alice's. His measurements performed in the $Z$ basis are identical to Alice's, as if Alice's measurements had influenced his (or vice versa).

\begin{equation*}
    \langle Z_AZ_B \rangle = \left( \frac{\bra{00} + \bra{11}}{\sqrt{2}} \right) \left( \frac{\ket{00} + (-1)^2 \ket{11}}{\sqrt{2}} \right) = 1
\end{equation*}

\begin{equation*}
    \langle Z_AX_B \rangle = \left( \frac{\bra{00} + \bra{11}}{\sqrt{2}} \right) \left( \frac{\ket{01} +(-1) \ket{10}}{\sqrt{2}} \right) = 0
\end{equation*}

The situation is very disturbing because it seems to contradict the principle of *locality*, which states that the evolution of physical states depends on their immediate environment, and not on what is at the other end of the universe.

In reality, this "paradox" calls into question the meaning of state superpositions and the very nature of quantum mechanics.

**EPR hypothesis (1935):** Quantum mechanics is incomplete. There would exist local hidden variables that predetermine measurement results.

```python
import numpy as np
from qiskit.quantum_info import Statevector, Operator
from qiskit.visualization import plot_histogram

psi_epr = Statevector([1, 0, 0, 1]) / np.sqrt(2)
display(psi_epr.draw("latex"))

Z = Operator.from_label("Z")
I = Operator.from_label("I")
X = Operator.from_label("X")

expectation_ZB = psi_epr.expectation_value(I ^ Z)
print(f"⟨Z_B⟩ = {expectation_ZB.real:.4f}")

expectation_XB = psi_epr.expectation_value(I ^ X)
print(f"⟨X_B⟩ = {expectation_XB.real:.4f}")

expectation_ZAZB = psi_epr.expectation_value(Z ^ Z)
print(f"⟨Z_A Z_B⟩ = {expectation_ZAZB.real:.4f}")

expectation_ZAXB = psi_epr.expectation_value(Z ^ X)
print(f"⟨Z_A X_B⟩ = {expectation_ZAXB.real:.4f}")
```

## Hidden variables

Defended idea: The probabilities $|\langle 0 | \psi \rangle|^2$ and $|\langle 1 | \psi \rangle|^2$ reflect an uncertainty about the nature of the state $\ket\psi$. There would be no superposition of states, but simply states whose exact nature we do not know.

A more fundamental theory than quantum mechanics would exactly describe the result of measuring an observable. Let us take for example the following superposition:

\begin{equation*}
    \ket\psi = \frac{\ket 0 + \ket 1}{\sqrt{2}}
\end{equation*}

The probability of measuring the value $+1$ for observable $Z$ is:

\begin{equation*}
    p(+1) = |\langle 0 | \psi \rangle|^2 = \frac 12
\end{equation*}

The hidden variables hypothesis assumes that the measurement result of $Z$, $z = \pm 1$, would be pre-determined and encoded in the quantum state. The quantum state, $\ket{\psi^{(z)}}$, would in reality be either $\ket{\psi^{(+1)}} = \ket 0$ or $\ket{\psi^{(-1)}} = \ket 1$, and the measurement would merely reveal the exact nature of our state. The probability $1/2$ of measuring $z=+1$ would then simply translate the probability that this hidden variable equals $+1$:

\begin{equation*}
    p(+1) = |\langle 0 | \psi^{(z)} \rangle|^2 = 
    \begin{cases}
        1 \quad \text{ if } z=+1 \\
        0 \quad \text{ if } z=-1
    \end{cases}
    \quad
    = p(z=+1)
\end{equation*}

The correlations observed for entangled states,

\begin{equation*}
    \ket{\psi_{AB}} = \frac{\ket{00} + \ket{11}}{\sqrt{2}}
\end{equation*}

\begin{equation*}
    \langle Z_AZ_B \rangle = \left( \frac{\bra{00} + \bra{11}}{\sqrt{2}} \right) \left( \frac{\ket{00} + (-1)^2 \ket{11}}{\sqrt{2}} \right) = 1
\end{equation*}

would also be explained by the pre-encoded values of the hidden variable:

\begin{equation*}
    \ket{\psi_{AB}} = %\ket{\psi_A^{(z)}}\otimes\ket{\psi_B^{(z)}}
    \begin{cases}
        \ket{\psi_A^{(+1)}}\otimes\ket{\psi_B^{(+1)}} \quad \text{ if } z=+1 \\
        \ket{\psi_A^{(-1)}}\otimes\ket{\psi_B^{(-1)}} \quad \text{ if } z=-1
    \end{cases}
\end{equation*}

\begin{equation*}
    \langle Z_AZ_B \rangle = \langle \psi_A^{(z)} | Z_A | \psi_A^{(z)} \rangle \langle \psi_B^{(z)} | Z_B | \psi_B^{(z)} \rangle = z^2 = 1
\end{equation*}

## Statistical test (CHSH)

<img src="../figures/chsh.png" alt="chsh" width="400"/>

The source emits states $\ket{\psi_A}$ and $\ket{\psi_B}$ toward Alice and Bob, respectively.

Alice decides to measure one observable chosen from $A_0, A_1$, and Bob does the same on his side with $B_0,B_1$.
Alice and Bob are sufficiently far apart so that one's result cannot influence the other's measurement.

The measurement results and chosen observables are collected and analyzed afterwards. We then compute the correlations $\langle A_iB_j \rangle$ and evaluate

\begin{equation*}
    S = \langle A_0B_0 \rangle + \langle A_0B_1 \rangle + \langle A_1B_0 \rangle - \langle A_1B_1 \rangle
    %= \langle A_0B_0 + A_0B_1 + A_1B_0 - A_1B_1 \rangle
\end{equation*}

If there exist hidden variables $a_0,a_1,b_0,b_1$ predetermining the measurement results of $A_0,A_1,B_0,B_1$, then the average combination $S$ will evaluate to

\begin{align*}
    S &= \sum_{a_0,a_1,b_0,b_1} p(a_0,a_1,b_0,b_1) \, \underbrace{(a_0b_0+ a_0b_1 + a_1b_0 - a_1b_1)}_{a_0(b_0+b_1) \, + \, a_1(b_0-b_1)}
\end{align*}

Since $b_0,b_1 = \pm 1$ we necessarily have one of the two factors $(b_0+b_1)$ or $(b_0-b_1)$ that vanishes, the other taking the value $\pm 2$. Thus,

\begin{equation*}
    a_0(b_0+b_1) + a_1(b_0-b_1) = \pm 2
\end{equation*}

and therefore

\begin{equation*}
    |S| \le 2
\end{equation*}

This equation is called the **CHSH inequality** and is an example of a Bell inequality. It serves to test the hypothesis of hidden variables.

## Quantum mechanics does not respect Bell inequalities

In the CHSH case, consider the following observables for Alice:

\begin{equation*}
    A_0 = Z
    \qquad
    A_1 = X
\end{equation*}

And these for Bob:

\begin{equation*}
    B_0 = \frac{Z+X}{2}
    \qquad
    B_1 = \frac{Z-X}{2}
\end{equation*}

If the source produces the states

\begin{equation*}
    \ket{\psi_{AB}} = \frac{\ket{01} - \ket{10}}{\sqrt{2}}
\end{equation*}

then the correlations will be

\begin{equation*}
    \langle A_0B_0 \rangle = +\frac{1}{\sqrt{2}} \qquad
    \langle A_0B_1 \rangle = +\frac{1}{\sqrt{2}} \qquad
    \langle A_1B_0 \rangle = +\frac{1}{\sqrt{2}} \qquad
    \langle A_1B_1 \rangle = -\frac{1}{\sqrt{2}}
\end{equation*}

And the combination $S = 2\sqrt{2}$

**Is quantum mechanics incomplete? Or is an interpretation in terms of local hidden variables that is impossible?**

```python
# Create the state for the CHSH test: (|01> - |10>)/sqrt(2)
psi_chsh = Statevector([0, 1, -1, 0]) / np.sqrt(2)
display(psi_chsh.draw("latex"))
```

```python
# Define observables for the CHSH test
# Alice: A_0 = Z, A_1 = X
A_0 = Operator.from_label("Z")
A_1 = Operator.from_label("X")

# Bob: B_0 = (Z+X)/sqrt(2), B_1 = (Z-X)/sqrt(2)
# We use rotations to implement these observables
Z_mat = np.array([[1, 0], [0, -1]])
X_mat = np.array([[0, 1], [1, 0]])

B_0_mat = (Z_mat + X_mat) / np.sqrt(2)
B_1_mat = (Z_mat - X_mat) / np.sqrt(2)

B_0 = Operator(B_0_mat)
B_1 = Operator(B_1_mat)

print("Alice's observables:")
display(A_0.draw("latex"))
display(A_1.draw("latex"))

print("Bob's observables:")
display(B_0.draw("latex"))
display(B_1.draw("latex"))
```

```python
# Compute correlations for the CHSH test
corr_A0B0 = psi_chsh.expectation_value(A_0 ^ B_0)
corr_A0B1 = psi_chsh.expectation_value(A_0 ^ B_1)
corr_A1B0 = psi_chsh.expectation_value(A_1 ^ B_0)
corr_A1B1 = psi_chsh.expectation_value(A_1 ^ B_1)

print(f"⟨A_0 B_0⟩ = {corr_A0B0.real:.4f}")
print(f"⟨A_0 B_1⟩ = {corr_A0B1.real:.4f}")
print(f"⟨A_1 B_0⟩ = {corr_A1B0.real:.4f}")
print(f"⟨A_1 B_1⟩ = {corr_A1B1.real:.4f}")

# Compute S
S = corr_A0B0 + corr_A0B1 + corr_A1B0 - corr_A1B1
print("\n" + "=" * 50)
print(f"S = {S.real:.4f}")
print("=" * 50)
print(f"\nViolation of the CHSH inequality: |S| = {abs(S.real):.4f} > 2")
print("Quantum mechanics is not compatible with the existence of local hidden variables!")
```
