# Quantum key distribution

```python
from qiskit.quantum_info import Statevector, Operator
from numpy import random

def print_formatted(data):
    formatted = '[' + ' '.join(str(x) for x in data) + ']'
    print(formatted)
```

**Alice:**

```python
N = 32
alice_data_bits = [int(x) for x in random.randint(0,2,N)]
alice_op_labels = random.choice(['H', 'I'], N).tolist()

print_formatted(alice_data_bits)
print_formatted(alice_op_labels)
```

```python
qubits = []
for i in range(N):
    data_bit = alice_data_bits[i]
    op_label = alice_op_labels[i]
    if op_label == 'I':
        if data_bit == 0:
            qubits.append(Statevector.from_label('0'))
        else:
            qubits.append(Statevector.from_label('1'))
    else:
        if data_bit == 0:
            qubits.append(Statevector.from_label('+'))
        else:
            qubits.append(Statevector.from_label('-'))

for i in range(5):
    display(qubits[i].draw('latex'))
```

**Alice:** I'm sending you my qubits...

```python
# Safe transmission, no one is spying
```

**Bob:** Received! I will also perform measurements!

```python
bob_op_labels = random.choice(['H', 'I'], N).tolist()
print_formatted(bob_op_labels)

bob_measurements = []

for i in range(N):
    Op = Operator.from_label(bob_op_labels[i])
    q = qubits[i].evolve(Op)

    outcome, state = q.measure()
    measured_bit = int(outcome)
    bob_measurements.append(measured_bit)

print_formatted(bob_measurements)
```

**Alice:** Here are the random operations I used

```python
print_formatted(alice_op_labels)
```

**Bob:** Received! Here are mine:

```python
print_formatted(bob_op_labels)
```

**Alice and Bob:** Let's select the qubits that underwent the same operations!

(These qubits are unchanged between encoding and reading, since $HH=I$)

```python
shared_data_bits = []

for i in range(N):
    if alice_op_labels[i] == bob_op_labels[i]:
        if alice_data_bits[i] == bob_measurements[i]:
            shared_data_bits.append(str(alice_data_bits[i]))
        else:
            shared_data_bits.append('x')
    else:
        shared_data_bits.append(' ')

print("Shared key:")
print_formatted(shared_data_bits)
```

Summary table:

```python
print_formatted(alice_data_bits)
print_formatted(alice_op_labels)
print_formatted(bob_op_labels)
print_formatted(bob_measurements)
print_formatted(shared_data_bits)
```

## Exercise: eavesdropping by a third party

Insert the code below in the middle of the qubit transfer between Alice and Bob, to observe the effect of eavesdropping by Eve.

**Eve:** 

```python
eve_op_labels = random.choice(['H', 'I'], N).tolist()
eve_measurements = []

for i in range(N):
    Op = Operator.from_label(eve_op_labels[i])
    q = qubits[i].evolve(Op)

    outcome, state = q.measure()
    measured_bit = int(outcome)
    eve_measurements.append(measured_bit)

    # Eve alters the qubit state by performing the measurement
    qubits[i] = state

print_formatted(eve_op_labels)
print_formatted(eve_measurements)
```

To address this issue, Alice and Bob verify the quality of the information transfer by comparing a portion of the shared bits. The remaining bits will constitute the secret key.

Finally, the communication of qubits is certainly imperfect, which will naturally lead to different keys, even without eavesdropping by a third party. To remedy this problem, there exist techniques for *information reconciliation* and *privacy amplification*.
