## What is a circuit?

![title](../figures/bell_circuit.png)

A computation to be executed on a quantum processor is typically represented as a circuit. The circuit is a schematic representation where qubits are represented by horizontal lines interrupted by *gates*, representing the operations performed on these qubits.

The circuit describes a general operation on all qubits, which are assumed to be initialized in state $\ket 0$, unless stated otherwise. It is read from left to right, with gates gradually modifying the quantum state of the system. The final state is therefore read at the output of the circuit (right). Physically, the gates represent operations that will actually be performed on the quantum state of the physical processor. The left-to-right reading can therefore be interpreted as a temporal evolution of the system.

To extract information from a quantum state, it is necessary to perform *measurements* on the qubits. Measurements are always assumed to be performed on all qubits of the final state, and are therefore placed on the right at the end of the horizontal lines. Measurements convert the information into a form interpretable by a human or classical computer, represented by the binary sequence of measurement results.

**A circuit in Qiskit is actually a much more complex object, which can also describe classical operations, loops, and conditional control flow**

## Building simple circuits

Creating a superposition state on a single qubit:

```python
from qiskit.circuit import QuantumCircuit
from qiskit.quantum_info import Statevector, Operator 

# 1: Create a 1-qubit circuit:
circuit = QuantumCircuit(1)

# 2: Add a Hadamard gate on qubit 0:
circuit.h(0)

# 3: Circuit visualization
display(circuit.draw('mpl'))

# 4: Visualization of the corresponding matrix:
circuit_matrix = Operator.from_circuit(circuit)
display(circuit_matrix.draw('latex'))

# 5: Visualization of the final state:
circuit_final_state = Statevector.from_circuit(circuit)
display(circuit_final_state.draw('latex'))
```

Bell pair:

```python
from qiskit.circuit import QuantumCircuit
from qiskit.quantum_info import Statevector, Operator 

# 1: Create a 2-qubit circuit:
circuit = QuantumCircuit(2)

# 2: Add a Hadamard gate on qubit 0 and a CNOT operation on 1 controlled by 0:
circuit.h(0)
circuit.cx(0,1)

# 3: Circuit visualization
display(circuit.draw('mpl'))

# 4: Visualization of the corresponding matrix:
circuit_matrix = Operator.from_circuit(circuit)
display(circuit_matrix.draw('latex'))

# 5: Visualization of the final state:
circuit_final_state = Statevector.from_circuit(circuit)
display(circuit_final_state.draw('latex'))
```

## Measurements and circuit simulation

```python
from qiskit.circuit import QuantumCircuit
from qiskit.quantum_info import Statevector, Operator 

circuit = QuantumCircuit(2)

circuit.h(0)
circuit.cx(0,1)

circuit.measure_all() ## measure all qubits simultaneously

display(circuit.draw('mpl'))
```

```python
from qiskit.primitives import StatevectorSampler

sampler = StatevectorSampler()
job = sampler.run([circuit], shots=1024) ## repeat the experiment 1024 times
result = job.result() ## retrieve the experiment results

from qiskit.visualization import plot_histogram
plot_histogram(result[0].data.meas.get_counts())
```
